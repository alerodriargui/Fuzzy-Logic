```python
# ===============================
# BLOQUE 1: Imports y universos
# ===============================
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Generate universe variables
perc_fatig = np.arange(0, 10.1, 0.1)  # Fatigue [0,10]
hr = np.arange(40, 91, 1)  # Heart Rate [40,90]
sleep_quality = np.arange(0, 10.1, 0.1)   # Sleep Quality [0,10]
intensity = np.arange(0, 101, 1)  # Exercise Intensity [0,100]

print("Universos creados:")
print("perc_fatig:", perc_fatig)
print("hr:", hr)
print("sleep_quality:", sleep_quality)
print("intensity:", intensity)

```

```python

# ===============================
# BLOQUE 2: membership functions --> fatigue
# ===============================

perc_fatig_very_low = fuzz.trimf(perc_fatig, [0, 0, 2])
perc_fatig_low = fuzz.trimf(perc_fatig, [0, 2, 4])
perc_fatig_medium = fuzz.trimf(perc_fatig, [2, 5, 8])
perc_fatig_high = fuzz.trimf(perc_fatig, [6, 8, 10])
perc_fatig_very_high = fuzz.trimf(perc_fatig, [8, 10, 10])
# plot perc_fatig_very_low
plt.figure()
plt.plot(perc_fatig, perc_fatig_very_low, 'b', linewidth=1.5, label='Very Low')
plt.plot(perc_fatig, perc_fatig_low, 'g', linewidth=1.5, label='Low')
plt.plot(perc_fatig , perc_fatig_medium, 'r', linewidth=1.5, label='Medium')
plt.plot(perc_fatig, perc_fatig_high, 'c', linewidth=1.5, label='High')
plt.plot(perc_fatig, perc_fatig_very_high,  'm', linewidth=1.5, label='Very High')
plt.title('Fatigue Level')
plt.xlabel('Fatigue Level')
plt.ylabel('Membership')
plt.legend()
plt.grid()
plt.show()
```

```python
# ===============================
# BLOQUE 2: membership functions --> HR
# ===============================

#hr
hr_low = fuzz.trimf(hr, [40, 40, 60])
hr_normal= fuzz.trimf(hr, [50, 65, 80])
hr_high = fuzz.trimf(hr, [70, 90, 90])
# plot hr_low
plt.figure()
plt.plot(hr, hr_low, 'b', linewidth=1.5, label='Low')
plt.plot(hr, hr_normal, 'g', linewidth=1.5, label='Normal')
plt.plot(hr, hr_high, 'r', linewidth=1.5, label='High')
plt.title('Heart Rate')
plt.xlabel('Heart Rate (bpm)')
plt.ylabel('Membership')
plt.legend()
plt.grid()
plt.show()
```

```python
# ===============================
# BLOQUE 2: membership functions --> Sleep
# ===============================

# Sleep Quality
sleep_quality_poor = fuzz.trimf(sleep_quality, [0, 0, 4])
sleep_quality_average = fuzz.trimf(sleep_quality, [2, 5, 8])
sleep_quality_good = fuzz.trimf(sleep_quality, [6, 10, 10])
# plot sleep_quality_poor
plt.figure()
plt.plot(sleep_quality, sleep_quality_poor, 'b', linewidth=1.5, label='Poor')
plt.plot(sleep_quality, sleep_quality_average, 'g', linewidth=1.5, label='Average')
plt.plot(sleep_quality, sleep_quality_good, 'r', linewidth=1.5, label='Good')
plt.title('Sleep Quality')
plt.xlabel('Sleep Quality')
plt.ylabel('Membership')
plt.legend()
plt.grid()
plt.show()
```

```python
# ===============================
# BLOQUE 2: membership functions --> Intensity
# ===============================

# Intensity
intensity_low = fuzz.trimf(intensity, [0, 0, 40])
intensity_low_medium = fuzz.trimf(intensity, [20, 40, 60])
intensity_medium = fuzz.trimf(intensity, [40, 60, 80])
intensity_medium_high = fuzz.trimf(intensity, [60, 80, 100])
intensity_high = fuzz.trimf(intensity, [80, 100, 100])

# plot intensity_low
plt.figure()
plt.plot(intensity, intensity_low, 'b', linewidth=1.5, label='Low')
plt.plot(intensity, intensity_low_medium, 'g', linewidth=1.5, label='Low-Medium')
plt.plot(intensity, intensity_medium, 'r', linewidth=1.5, label='Medium')
plt.plot(intensity, intensity_medium_high, 'c', linewidth=1.5, label='Medium-High')
plt.plot(intensity, intensity_high, 'm', linewidth=1.5, label='High')
plt.title('Exercise Intensity')
plt.xlabel('Intensity (%)')
plt.ylabel('Membership')
plt.legend()
plt.grid()
plt.show()
```

```python
# ===============================
# BLOQUE 3: Fuzzy rules
# ===============================

# entry values
fatigue_input = 3
hr_input = 65     
sleep_input = 6

# -------------------------------
# 1 Evaluate membership for the entries
# -------------------------------
# Cansancio percibido
fatig_very_low_lvl = fuzz.interp_membership(perc_fatig, perc_fatig_very_low, fatigue_input)
fatig_low_lvl = fuzz.interp_membership(perc_fatig, perc_fatig_low, fatigue_input)
fatig_med_lvl = fuzz.interp_membership(perc_fatig, perc_fatig_medium, fatigue_input)
fatig_high_lvl = fuzz.interp_membership(perc_fatig, perc_fatig_high, fatigue_input)
fatig_vhigh_lvl = fuzz.interp_membership(perc_fatig, perc_fatig_very_high, fatigue_input)

# HR
hr_low_lvl = fuzz.interp_membership(hr, hr_low, hr_input)
hr_normal_lvl = fuzz.interp_membership(hr, hr_normal, hr_input)
hr_high_lvl = fuzz.interp_membership(hr, hr_high, hr_input)

# Sueo
sleep_poor_lvl = fuzz.interp_membership(sleep_quality, sleep_quality_poor, sleep_input)
sleep_avg_lvl = fuzz.interp_membership(sleep_quality, sleep_quality_average, sleep_input)
sleep_good_lvl = fuzz.interp_membership(sleep_quality, sleep_quality_good, sleep_input)

# Inicializar intensidad base
intensity_zero = np.zeros_like(intensity)

# -------------------------------
# 2 apply the rules
# -------------------------------

# Helper: OR = np.fmax(), AND = np.fmin()
# Regla 1: Cansancio bajo, HR baja, Sueo bueno  intensidad alta
rule1 = np.fmin(np.fmin(fatig_low_lvl, hr_low_lvl), sleep_good_lvl)
intensity_activation_high_1 = np.fmin(rule1, intensity_high)

# Regla 2: Cansancio bajo, HR baja, Sueo regular  intensidad media alta
rule2 = np.fmin(np.fmin(fatig_low_lvl, hr_low_lvl), sleep_avg_lvl)
intensity_activation_mhigh_2 = np.fmin(rule2, intensity_medium_high)

# Regla 3: Cansancio bajo, HR normal, Sueo bueno  intensidad media alta
rule3 = np.fmin(np.fmin(fatig_low_lvl, hr_normal_lvl), sleep_good_lvl)
intensity_activation_mhigh_3 = np.fmin(rule3, intensity_medium_high)

# Regla 4: Cansancio medio, HR normal, Sueo regular  intensidad media
rule4 = np.fmin(np.fmin(fatig_med_lvl, hr_normal_lvl), sleep_avg_lvl)
intensity_activation_med_4 = np.fmin(rule4, intensity_medium)

# Regla 5: Cansancio medio, HR alta, Sueo regular  intensidad media baja
rule5 = np.fmin(np.fmin(fatig_med_lvl, hr_high_lvl), sleep_avg_lvl)
intensity_activation_mlow_5 = np.fmin(rule5, intensity_low_medium)

# Regla 6: Cansancio alto, HR alta, Sueo malo  intensidad baja
rule6 = np.fmin(np.fmin(fatig_high_lvl, hr_high_lvl), sleep_poor_lvl)
intensity_activation_low_6 = np.fmin(rule6, intensity_low)

# Regla 7: Cansancio medio, HR normal, Sueo bueno  intensidad media alta
rule7 = np.fmin(np.fmin(fatig_med_lvl, hr_normal_lvl), sleep_good_lvl)
intensity_activation_mhigh_7 = np.fmin(rule7, intensity_medium_high)

# Regla 8: Cansancio bajo, HR alta, Sueo bueno  intensidad media
rule8 = np.fmin(np.fmin(fatig_low_lvl, hr_high_lvl), sleep_good_lvl)
intensity_activation_med_8 = np.fmin(rule8, intensity_medium)

# Regla 9: Cansancio alto, HR baja, Sueo bueno  intensidad media
rule9 = np.fmin(np.fmin(fatig_high_lvl, hr_low_lvl), sleep_good_lvl)
intensity_activation_med_9 = np.fmin(rule9, intensity_medium)

# Regla 10: Cansancio alto, HR normal, Sueo regular  intensidad media baja
rule10 = np.fmin(np.fmin(fatig_high_lvl, hr_normal_lvl), sleep_avg_lvl)
intensity_activation_mlow_10 = np.fmin(rule10, intensity_low_medium)

# Regla 11: Cansancio bajo, HR normal, Sueo malo  intensidad media baja
rule11 = np.fmin(np.fmin(fatig_low_lvl, hr_normal_lvl), sleep_poor_lvl)
intensity_activation_mlow_11 = np.fmin(rule11, intensity_low_medium)

# Regla 12: Cansancio medio, HR baja, Sueo bueno  intensidad media alta
rule12 = np.fmin(np.fmin(fatig_med_lvl, hr_low_lvl), sleep_good_lvl)
intensity_activation_mhigh_12 = np.fmin(rule12, intensity_medium_high)

# -------------------------------
# 3 add all activations
# -------------------------------
aggregated = np.fmax(
    intensity_activation_high_1,
    np.fmax(
        intensity_activation_mhigh_2,
        np.fmax(
            intensity_activation_mhigh_3,
            np.fmax(
                intensity_activation_med_4,
                np.fmax(
                    intensity_activation_mlow_5,
                    np.fmax(
                        intensity_activation_low_6,
                        np.fmax(
                            intensity_activation_mhigh_7,
                            np.fmax(
                                intensity_activation_med_8,
                                np.fmax(
                                    intensity_activation_med_9,
                                    np.fmax(
                                        intensity_activation_mlow_10,
                                        np.fmax(
                                            intensity_activation_mlow_11,
                                            intensity_activation_mhigh_12
                                        ),
                                    ),
                                ),
                            ),
                        ),
                    ),
                ),
            ),
        ),
    ),
)

# -------------------------------
# 4 output
# -------------------------------
plt.figure(figsize=(8, 3))
plt.plot(intensity, intensity_low, 'b--', linewidth=0.5)
plt.plot(intensity, intensity_medium_high, 'r--', linewidth=0.5)
plt.fill_between(intensity, intensity_zero, aggregated, facecolor='orange', alpha=0.7)
plt.title('Aggregated Output Intensity')
plt.xlabel('Exercise Intensity (%)')
plt.ylabel('Membership')
plt.grid()
plt.show()
```

```python
# ===============================
# BLOQUE 4: Defuzzification
# ===============================

# (centroid)
intensity_result = fuzz.defuzz(intensity, aggregated, 'centroid')

# degree of relevance
activation = fuzz.interp_membership(intensity, aggregated, intensity_result)

# print results
print(f" Intensidad final recomendada: {intensity_result:.2f} %")

# Visualizacin
plt.figure(figsize=(8, 3))
plt.plot(intensity, intensity_low, 'b', linewidth=0.5, linestyle='--')
plt.plot(intensity, intensity_medium_high, 'r', linewidth=0.5, linestyle='--')
plt.fill_between(intensity, intensity_zero, aggregated, facecolor='orange', alpha=0.7)
plt.plot([intensity_result, intensity_result], [0, activation], 'k', linewidth=1.5, linestyle='--')
plt.title('Defuzzificacin del resultado')
plt.xlabel('Exercise Intensity (%)')
plt.ylabel('Membership')
plt.grid()
plt.show()
```
